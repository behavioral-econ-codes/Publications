```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


def PlotTopicsNumTesting(res_dir, var_x_str, var_y_str):
    gr_dir='.\\graphs_results\\'
    
    models_file='models_testing.csv'

    header=['num','topics_num','class_ids','steps_num','strategy','perplexity','Phi-sparcity','Theta-sparcity','Kernel_size','Contrast','Purity','Coherence']
    test_df = pd.read_csv(res_dir+models_file, delimiter=',', names=header)
    pd.options.display.float_format = '{:,.4f}'.format
    
    test_df = test_df.iloc[1:]
    test_df=test_df.iloc[127:151]
    #test_df2=test_df.iloc[36:42]
    #test_df=pd.concat([test_df1, test_df2])
    
    
    test_df['topics_num']=test_df['topics_num'].astype("float")
    test_df['Coherence']=test_df['Coherence'].astype("float")
    test_df['perplexity']=test_df['perplexity'].astype("float")
    test_df['Kernel_size']=test_df['Kernel_size'].astype("float")
    test_df['Phi-sparcity']=test_df['Phi-sparcity'].astype("float")
    test_df['Contrast']=test_df['Contrast'].astype("float")
    test_df['Purity']=test_df['Purity'].astype("float")
#     test_df['strategy'] = test_df['strategy'].map({'[DecorrPhi]: ; 5.00e-02; ': 'tau=1.00e-02',
#                                                 '[DecorrPhi]: ; 5.00e-02; ': 'tau=5.00e-02', 
#                                                '[DecorrPhi]: ; 1.00e-01; ': 'tau=1.00e-01',
#                                                '[DecorrPhi]: ; 1.50e-01; ': 'tau=1.50e-01',
#                                                '[DecorrPhi]: ; 2.00e-01; ': 'tau=2.00e-01'
#                                               })

    fig, g = plt.subplots(figsize=(5,4))
    
    g=sns.scatterplot(data=test_df, x=var_x_str, y=var_y_str, hue='strategy', s=50, marker='x')
    g.set(xticks=[5, 10, 15, 20])

    legend = g.legend(loc='center left', bbox_to_anchor=(1.05, 0.15), ncol=1)
    legend.texts[0].set_text("DecorrPhiRegularizer")
    plt.grid()
    for label in (g.get_xticklabels() + g.get_yticklabels()):
        label.set_fontname('Times New Roman')
        label.set_fontsize(12)
    
    g.set_xlabel(u'Число тем', fontname='Times New Roman', fontsize=12)
    g.set_ylabel(var_y_str, fontname='Times New Roman', fontsize=12)

    plt.title(u'Тестирование числа тем', fontsize=15, fontname='Times New Roman', x=0.5, y=1.01)
    name_png=var_x_str+ '_'+var_y_str+'modality.png'
    #plt.savefig(gr_dir+name_png, bbox_inches = 'tight')
```

```python
res_dir='.\\final_tests4\\'

PlotTopicsNumTesting(res_dir, 'topics_num', 'Coherence')
PlotTopicsNumTesting(res_dir, 'topics_num', 'perplexity')
PlotTopicsNumTesting(res_dir, 'topics_num', 'Kernel_size')
```

```python
# Тестирование регуляризаторов
import seaborn as sns
import matplotlib.pyplot as plt


def PlotRegularizersTesting(res_dir, steps_num, var_x_str, var_y_str):
    gr_dir='.\\graphs_results\\'
    
    models_file='models_testing.csv'

    header=['num','topics_num','class_ids','steps_num','strategy','perplexity','Phi-sparcity','Theta-sparcity','Kernel_size','Contrast','Purity','Coherence']
    test_df = pd.read_csv(res_dir+models_file, delimiter=',', names=header)
    #pd.options.display.float_format = '{:,.4f}'.format
    
#     test_df.class_ids= test_df.class_ids.str.replace('{','',regex=True)
#     test_df.class_ids= test_df.class_ids.str.replace('}','',regex=True)
#     test_df.class_ids= test_df.class_ids.str.replace('u\'', '', regex=False)
#     test_df.class_ids= test_df.class_ids.str.replace('\'','',regex=True)
    
    
    test_df1 = test_df.iloc[158:163]
    
    test_df2=test_df.iloc[167:170]
    
    test_df=pd.concat([test_df1, test_df2])
    
    
    test_df['topics_num']=test_df['topics_num'].astype("int")
    test_df['num']=test_df['num'].astype("float")
    test_df['steps_num']=test_df['steps_num'].astype("float")
    
    test_df['Coherence']=test_df['Coherence'].astype("float")
    test_df['perplexity']=test_df['perplexity'].astype("float")
    test_df['Kernel_size']=test_df['Kernel_size'].astype("float")
    test_df['Phi-sparcity']=test_df['Phi-sparcity'].astype("float")
    test_df['Contrast']=test_df['Contrast'].astype("float")
    test_df['Purity']=test_df['Purity'].astype("float")
    
    
#     test_df['class_ids'] = test_df['class_ids'].map({
#                                                'bigrams: 1.0, authors: 2.0': 'bigrams: 1.0; authors: 2.0', 
#                                                'bigrams: 1.0, references: 1.0, authors: 2.0': 'bigrams: 1.0; authors: 2.0; references: 1.0'
                                               
#                                               })
    test_df_step=test_df.loc[test_df.steps_num==steps_num]  #.iloc[5:]
    #test_df_step=test_df_step.iloc[35:]
    test_df_step_=test_df_step.reset_index()
    test_df_step_.index = test_df_step_.index + 1
#     print test_df 
    fig, g = plt.subplots(figsize=(5,4))
    
    g=sns.scatterplot(data=test_df_step_, x=test_df_step_.index, y=var_y_str, s=50, hue='strategy', marker='x') #x=test_df_step_.index
    g.set(xticks=[1, 2, 3, 4, 5, 6, 7, 8])

    legend = g.legend(loc='center left', bbox_to_anchor=(1.05, 0.15), ncol=1)
    legend.texts[0].set_text(u"Значения коэффициентов регуляризации")
    plt.grid()
    for label in (g.get_xticklabels() + g.get_yticklabels()):
        label.set_fontname('Times New Roman')
        label.set_fontsize(12)
    
    g.set_xlabel(u'', fontname='Times New Roman', fontsize=12)
    g.set_ylabel(var_y_str, fontname='Times New Roman', fontsize=12)

    plt.title(u'Тестирование коэффициентов разреживания', fontsize=15, fontname='Times New Roman', x=0.5, y=1.04)
    name_png=var_x_str+ '_'+var_y_str+'_sparse'+'.png'
    #plt.savefig(gr_dir+name_png, bbox_inches = 'tight')
```

```python
res_dir='.\\final_tests4\\'


PlotRegularizersTesting(res_dir, 20, 'topics_num', 'Coherence')
PlotRegularizersTesting(res_dir, 20, 'topics_num', 'perplexity')
```
